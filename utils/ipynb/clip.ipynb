```python
import numpy as np
import matplotlib.pyplot as plt

def plot_multi_columns(txt_path):
    # 读取数据，跳过表头
    data = np.loadtxt(txt_path, skiprows=1)
    time = data[:, 0]
    poutout = data[:, 1]
    iz1 = data[:, 2]
    vin = data[:, 3]

    # 配置线型
    line_style = dict(linestyle='-', linewidth=1)

    # 绘制 Poutout vs time
    plt.figure(figsize=(16, 8))
    plt.plot(time, poutout, **line_style)
    plt.xlabel('time')
    plt.ylabel('Poutout')
    plt.title('Poutout vs time')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 绘制 Iz1 vs time
    plt.figure(figsize=(16, 8))
    plt.plot(time, iz1, **line_style)
    plt.xlabel('time')
    plt.ylabel('Iz1')
    plt.title('Iz1 vs time')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 绘制 Vin vs time
    plt.figure(figsize=(16, 8))
    plt.plot(time, vin, **line_style)
    plt.xlabel('time')
    plt.ylabel('Vin')
    plt.title('Vin vs time')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# 示例用法
plot_multi_columns(r"D:\UNIPIC\Unipic2.5D_Training\UNIPIC20240819\bin\pic\TM03_60Ghz\Simulation\result\FieldsDgn.txt")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sequential(txt_path):
    # 读取数据，跳过标题行
    data = np.loadtxt(txt_path, skiprows=1)
    time = data[:, 0]
    poutout = data[:, 1]
    index = np.arange(1, len(time)+1)  # 横轴为1,2,3,...
    
    # 绘制time随序号变化曲线
    plt.figure(figsize=(8, 4))
    plt.plot(index, time, marker='o')
    plt.xlabel('Index')
    plt.ylabel('time')
    plt.title('time vs. Index')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 绘制Poutout随序号变化曲线
    plt.figure(figsize=(8, 4))
    plt.plot(index, poutout, marker='o')
    plt.xlabel('Index')
    plt.ylabel('Poutout')
    plt.title('Poutout vs. Index')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# 示例用法
plot_sequential(r"D:\UNIPIC\Unipic2.5D_Training\UNIPIC20240819\bin\pic\TM03_60Ghz\Simulation\result\FieldsDgn.txt")
```
